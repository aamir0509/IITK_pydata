```
# closest to origin

coordinates = np.random.randn(100, 2)
dist=[]
for coordinate in coordinates:
  res=np.sqrt(coordinate[0]**2+coordinate[1]**2)
  dist.extend([res])

top_10=np.sort(dist)[:10]
#print(top_10)
for cor in top_10:
  print(coordinates[dist.index(cor),:])
```

```
# circle redius

coordinates = np.random.randn(100, 2)
dist=[]
for coordinate in coordinates:
  res=np.sqrt(coordinate[0]**2+coordinate[1]**2)
  dist.extend([res])

radius_circle=int(input('enter the raduis of circle'))
for distance in dist:
  if distance<radius_circle:
    print(coordinates[dist.index(distance),:])
```

```
# calculte the pi

inside_circle = 0
total_points = 10000000

for i in range(0, total_points):
  x = np.random.uniform(-1,1)
  y = np.random.uniform(-1,1)
  if np.sqrt(x**2 + y**2) < 1.0:
      inside_circle += 1
pi = (float(inside_circle) / total_points) * 4
print(pi)
```

```
# calculte the pi short code

total_points = 10000000
x = np.random.uniform(-1,1,total_points)
y = np.random.uniform(-1,1,total_points)
new=np.sqrt(np.add(x**2 , y**2))
inside=new[new<1.0].shape

pi = (float(inside[0]) / total_points) * 4
print(pi)
```
